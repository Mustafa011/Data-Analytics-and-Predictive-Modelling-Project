```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

# Step 1: Load the dataset (replace with your actual dataset path)
data = pd.read_csv("encoded_dataset.csv")

# Step 2: Select numerical variables for the analysis
numerical_features = data[["Age", "Height", "Weight", "TUE"]]

# Step 3: Normalize the features (optional, but recommended)
normalized_features = (numerical_features - numerical_features.mean()) / numerical_features.std()

# Step 4: Compute pairwise weighted dissimilarities (e.g., using Euclidean distance)
distance_matrix = squareform(pdist(normalized_features, metric='euclidean'))

# Step 5: Compute the weighted dissimilarity matrix (4x4)
weights = np.array([1, 1, 1, 1])  # Assuming equal weights for simplicity
weighted_matrix = distance_matrix * weights[:, np.newaxis]
weighted_matrix = weighted_matrix[:4, :4]  # Restricting to 4x4 matrix for visualization

# Step 6: Visualize the weighted dissimilarity matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(weighted_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("4x4 Weighted Dissimilarity Matrix")
plt.xlabel("Features")
plt.ylabel("Features")
plt.xticks(ticks=np.arange(4) + 0.5, labels=["Age", "Height", "Weight", "TUE"])
plt.yticks(ticks=np.arange(4) + 0.5, labels=["Age", "Height", "Weight", "TUE"], rotation=0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

# Step 1: Load the dataset
data = pd.read_csv("encoded_dataset.csv")  # Replace with your dataset path

# Step 2: Select numerical variables (Age, Height, Weight, TUE)
numerical_data = data[["Age", "Height", "Weight", "TUE"]]

# Step 3: Define weights for each variable
weights = np.array([1, 1, 1, 1])  # Replace with specific weights if needed

# Step 4: Apply weights to the dataset
weighted_data = numerical_data * weights

# Step 5: Compute the pairwise distances (Euclidean distance)
distance_matrix = squareform(pdist(weighted_data, metric='euclidean'))

# Step 6: Normalize the distance matrix to range [0, 2]
normalized_matrix = 2 * (distance_matrix - np.min(distance_matrix)) / (np.max(distance_matrix) - np.min(distance_matrix))

# Step 7: Visualize the weighted dissimilarity matrix (4x4) as a heatmap with annotations
plt.figure(figsize=(8, 6))
plt.title("Weighted Dissimilarity Matrix (4x4)")
heatmap = plt.imshow(normalized_matrix[:4, :4], cmap="coolwarm", interpolation="nearest")

# Adding annotations
for i in range(4):
    for j in range(4):
        plt.text(j, i, f"{normalized_matrix[i, j]:.2f}", ha="center", va="center", color="black")

# Adding color bar and labels
plt.colorbar(heatmap, label="Weighted Distance")
plt.xticks(range(4), ["Age", "Height", "Weight", "TUE"])
plt.yticks(range(4), ["Age", "Height", "Weight", "TUE"])
plt.show()
```
